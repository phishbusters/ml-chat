#### Palabras clave

En todo mensaje, hay palabras clave que podemos encontrar.
Esta pequena lista ayuda a que esas palabras clave puedan ser detectadas.
Esta para modelos que no son como LSTM

```python
key_words = [
    'gana', 
    'gratis', 
    'cuenta', 
    'segura',
    'datos', 
    'dni', 
    'pasaporte', 
    'domicilio', 
    'credito', 
    'tarjeta', 
    'debito',
    'codigo de seguridad',
    'numero de tramite',
    'transferime',
    'transfiere',
    'pin'
]
```

### Creacion del archivo

Aqui solo combinamos los array anteriores y creamos un CSV.

```python
def extract_url_features(message):
    # Extraer la URL
    url = re.findall(r'(?:https?://|www\.)\S+', message)  # Buscar una URL que comience con http, https o www
    if url:
        url = url[0]
        # Agregar "http://" si la URL comienza con "www." para que urlparse funcione correctamente
        if url.startswith("www."):
            url = "http://" + url
        parsed_url = urlparse(url)
        return pd.Series({
            'url': url,
            'dominio': parsed_url.netloc,
            'tiene_https': 1 if parsed_url.scheme == 'https' else 0,
            'longitud_url': len(url),
            'presencia_subdominio': 1 if '.' in parsed_url.netloc.split('.')[0] else 0,
            'palabras_clave_sospechosas': 1 if any(keyword in message for keyword in key_words) else 0
        })
    else:
        return pd.Series({
            'url': '',
            'dominio': '',
            'tiene_https': 0,
            'longitud_url': 0,
            'presencia_subdominio': 0,
            'palabras_clave_sospechosas': 1 if any(keyword in message for keyword in key_words) else 0
        })


phishing_messages = phishing_real_cases +  frases_argentas_phishing + others_phishing
no_phishing_messages = non_phishing_real_cases + frases_argentas_no_phishing + others_non_phishing
is_phishing = [1]*len(phishing_messages) + [0]*len(no_phishing_messages)

# Creando el DataFrame con los mensajes generados
data = {
    'message': phishing_messages + no_phishing_messages,
    'phishing': is_phishing
}
df = pd.DataFrame(data)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df = df.join(df['message'].apply(extract_url_features))

file_path = f"dataset_phishing_{datetime.today().strftime('%Y-%m-%d')}.csv"
df.to_csv(file_path, sep='\t', index=False)

# Ver las primeras 10 filas del dataset
df.head(1000)


```
